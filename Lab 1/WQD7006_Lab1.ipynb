## Part 2: Classification – Weather Prediction

```python
data = {
    'humidity': [30, 70, 90, 40, 80, 85],
    'temperature': [32, 28, 25, 34, 26, 24],
    'weather': ['Sunny', 'Cloudy', 'Rainy', 'Sunny', 'Rainy', 'Rainy']
}

df = pd.DataFrame(data)
df
```

```python
# labelEncoder or one hot encoder?
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['weather_encoded'] = le.fit_transform(df['weather'])
df
```

```python
# Simple example using decision tree
from sklearn.tree import DecisionTreeClassifier

X = df[['humidity', 'temperature']]
y = df['weather_encoded']

model = DecisionTreeClassifier(random_state=0)
model.fit(X, y)

pred = model.predict(X)
accuracy_score(y, pred)
```

**Discussion:**
- Why is this a classification problem?
- What is the performance measure used?

## Part 4: Unsupervised Learning – Patient Clustering

```python
from sklearn.cluster import KMeans

patient_data = np.array([
    [80, 2], [85, 3], [60, 1],
    [30, 5], [25, 4], [40, 6]
])

kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit_predict(patient_data)

plt.scatter(patient_data[:,0], patient_data[:,1], c=clusters)
plt.xlabel('Effectiveness')
plt.ylabel('Side Effects')
plt.title('Patient Response Clusters')
plt.show()
```

**Discussion:**
- Why is this unsupervised learning?
- How can this help medical decision making?
